# A+H Premium Regression

**Status**: stub — no spec phase implements this; kept as an ad-hoc analysis slot.

## Why this notebook still exists

Unlike notebooks 01 / 02 / 04 / 06 (deleted in the post-v1.0 cleanup because
their intended functions were absorbed into formal modules — Phase 2 ETL /
Phase 5 `cornerstone_signal_agent` / Phase 9 `docs/case_studies/` / Phase 8c
`backtest/calibration.py` respectively), A+H premium regression has **no
equivalent in the formal codebase**. It is genuinely missing analysis surface.

## How to start when you pick this up

1. Pull A-share / H-share paired data from `ipo_events` + `ipo_postmarket`
   (already ETL'd from NACS SQLite per ADR 0005 §1). Filter for A+H dual
   listings only.
2. Compute premium = `(A_price - H_price_hkd_equivalent) / H_price_hkd_equivalent`
   over the post-listing window (T+1 / T+30 / T+90 / T+180 — same checkpoints as
   `prediction_outcomes`).
3. Regress premium against candidate features: HSI volatility / Southbound flow
   / sector / market cap / float. `data/knowledge_base/market_env_cache.json`
   has the regime signal already.
4. If results are stable, lift the model into `valuation/ah_premium.py` and
   plumb it into `valuation/ensemble.py` as an additional adjustment for A+H
   names. Otherwise leave findings here.
