# Sponsor Track Record (analysis layer)

**Status**: stub — no spec phase implements the analysis layer; kept as an
ad-hoc analysis slot.

## Why this notebook still exists

Phase 2 implemented the **data layer** at `src/hk_ipo_agent/data/builders/
sponsor_track_record.py` (builds the per-sponsor aggregate table). What it does
**not** implement is the **analysis layer** — i.e. the actual question:
*"Does sponsor tier correlate with realised post-listing returns or with
decision accuracy in our prediction snapshots?"* That is the question this
notebook is meant to answer when someone gets around to it.

## How to start when you pick this up

1. Load the sponsor track record table built by
   `data/builders/sponsor_track_record.py`. Validate row count and date
   coverage first.
2. Join against `prediction_snapshots` (Phase 7.5) + `prediction_outcomes`
   (Phase 7.5b) to get realised 30/90/180-day returns per sponsor.
3. Slice by sponsor tier (1 = bulge-bracket / 2 = mid-tier / 3 = boutique) and
   test: does Tier-1 sponsorship systematically reduce downside vs. Tier-3?
   Use ADR 0013's `metrics.py` (Rank IC + L-S spread + Welch t-stat) instead
   of rolling your own statistics.
4. If a signal exists, lift it into `agents/fundamental_agent.py` or
   `valuation/ensemble.py` as a sponsor-quality adjustment. Otherwise document
   findings here.

Source-of-truth references:
- ADR 0013 §metrics (statistical primitives)
- `src/hk_ipo_agent/backtest/metrics.py` (Rank IC / L-S / t-stat — use these,
  don't reinvent)
- `src/hk_ipo_agent/data/builders/sponsor_track_record.py` (data source)
- `tests/unit/backtest/test_as_of_data.py` for as-of-date discipline (analysis
  using post-listing returns must respect this).
